```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])
```

```python
df
```

```python
X = df.iloc[ : ,[3,4]].values
X
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=3,min_samples=4,metric='euclidean')
```

```python
model = db.fit_predict(X)
model
```

```python
labels =db.labels_
print(labels)
```

```python
n_cluster = len(set(labels)) - (1 if -1 in labels else 0)
print(n_cluster)
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(X[labels == 0, 0], X[labels == 0, 1], color='yellow',label='cluster 0', s=50)
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], color='grey', label='cluster 1', s=50)
plt.scatter(X[labels == 2, 0], X[labels == 2, 1], color='blue', label='cluster 2', s=50)
plt.scatter(X[labels == 3, 0], X[labels == 3, 1], color='orange', label='cluster 3', s=50)
plt.scatter(X[labels == 4, 0], X[labels == 4, 1], color='green', label='cluster 4', s=50)
plt.scatter(X[labels == 5, 0], X[labels == 5, 1], color='red', label='cluster 5', s=50)
plt.scatter(X[labels == 6, 0], X[labels == 6, 1], color='violet', label='cluster 6', s=50)
plt.scatter(X[labels == 7, 0], X[labels == 7, 1], color='black', label='cluster 7', s=50)
plt.scatter(X[labels == 8, 0], X[labels == 8, 1], color='white', label='cluster 8', s=50)
plt.grid(False)
plt.title('DBSCAN')
plt.xlabel('Annual_Income_(k$)')
plt.ylabel('Spending_Score')
plt.legend()
plt.show()
```
